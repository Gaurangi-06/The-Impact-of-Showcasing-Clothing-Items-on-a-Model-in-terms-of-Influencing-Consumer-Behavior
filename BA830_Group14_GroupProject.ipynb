```python
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer
```

```python
df_regression = df_new.copy()
df_regression['Residence'] = df_regression['Residence'].replace({'Taiwan':1,'USA':2,'India':3,'Netherlands':4,'Canada':5,'Columbia':6, 'UK':7})
df_regression.rename(columns={'Age Group':'Age_group'},inplace=True)
df_regression
```

```python
df_regression['Gender'] = df_regression.Gender.astype('category')
df_regression['Age_group'] = df_regression.Age_group.astype('category')
df_regression['Cost'] = df_regression.Cost.astype('category')
df_regression['Residence'] = df_regression.Residence.astype('category')
df_regression['Industry'] = df_regression.Industry.astype('category')
```

```python
reg = smf.ols('Value ~ Treatment',data=df_regression).fit(cov_type='HC1')
Stargazer([reg])
```

```python
# Without interactions
reg_1 = smf.ols('Value ~ Treatment + Gender',data=df_regression).fit(cov_type='HC1')
reg_2 = smf.ols('Value ~ Treatment + Age_group',data=df_regression).fit(cov_type='HC1')
reg_3 = smf.ols('Value ~ Treatment + Residence',data=df_regression).fit(cov_type='HC1')
reg_4 = smf.ols('Value ~ Treatment + Cost',data=df_regression).fit(cov_type='HC1')
Stargazer([reg,reg_1,reg_2,reg_3,reg_4])
```

```python
# With interactions
reg_1 = smf.ols('Value ~ Treatment + Gender + Treatment*Gender',data=df_regression).fit(cov_type='HC1')
reg_2 = smf.ols('Value ~ Treatment + Age_group + Treatment*Age_group',data=df_regression).fit(cov_type='HC1')
reg_3 = smf.ols('Value ~ Treatment + Residence + Treatment*Residence',data=df_regression).fit(cov_type='HC1')
reg_4 = smf.ols('Value ~ Treatment + Cost + Treatment*Cost',data=df_regression).fit(cov_type='HC1')
Stargazer([reg,reg_1,reg_2,reg_3,reg_4])
```

### Randomization Check

```python
from statsmodels.stats.proportion import proportions_ztest

t_group = df_new[df_new['Treatment']==1]
c_group = df_new[df_new['Treatment']==0]
t_ate = df_new[df_new['Treatment'] == 1]['Value'].mean()
c_ate = df_new[df_new['Treatment'] == 0]['Value'].mean()

_, p_val = proportions_ztest(len(t_group), len(t_group+c_group) , value = 0.5)
print(p_val)
```

```python
from statsmodels.stats.power import TTestPower
from pingouin import power_ttest2n
```

```python
# Cohens D - method 1
t_std = df_regression[df_regression['Treatment'] == 1]['Value'].std()
c_std = df_regression[df_regression['Treatment'] == 0]['Value'].std()
sd = (((len(t_group)-1) * (t_std*2) + (len(c_group)-1) * (c_std*2)) / (len(t_group) + len(c_group)-2))*(1/2)
cohens_d = (t_ate-c_ate)/sd
print(cohens_d)
```

```python
# Cohens D - method 2
std_conversion = df_regression['Value'].std()

cohens_d = (t_ate-c_ate) / std_conversion
print('Cohens D:', cohens_d)
```

```python
# Detecting Power
power = power_ttest2n(nx = len(t_group), ny = len(c_group), d = cohens_d,
                      alpha = 0.05, alternative = 'two-sided')
power
```

```python
# Sample_size
n = TTestPower().solve_power(effect_size = cohens_d,
                            nobs = None, alpha = .05, power = 0.8, alternative='two-sided')

print(f"Necessary sample size: {np.ceil(n)}")
```
